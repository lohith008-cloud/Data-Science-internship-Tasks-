## Q2.Given an integer, , perform the following conditional actions:
        # If n is odd, print Weird
        # If n is even and in the inclusive range of 2 to 5, print Not Weird
        # If n is even and in the inclusive range of 6 to 20 , print Weird
        # If n is even and greater than 20 , print Not Weird

```python
number = int(input())

if number % 2 != 0:
    print("Weird")
else:
    if 2 <= number <= 5:
        print("Not Weird")
    elif 6 <= number <= 20:
        print("Weird")
    else:
        print("Not Weird")
```

## Q6. Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean True, otherwise return False.

Note that the code stub provided reads from STDIN and passes arguments to the is_leap function. It is only necessary to complete the is_leap function.

```python
def is_leap(year_value):
    leap = False
    
    if year_value % 4 == 0:
        if year_value % 100 != 0 or year_value % 400 == 0:
            leap = True

    return leap

yr = int(input())
print(is_leap(yr))
```
